```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
A = np.random.rand(50, 50) + 0.5
```

```python
A = np.tril(A)
```

```python
plt.imshow(A)
```

```python
np.linalg.det(A)
```

```python
plt.stem(A @ x)
```

```python
x = np.ones(50)
```

```python
z0 = A @ x
z = z0
```

```python
J = np.diag(A)
alpha = 0.5
```

```python
plt.figure(figsize=(12, 3))
for i in range(61):
    z = z - alpha * np.diag(1 / J) @ (A @ z - z0)
    if i % 20 == 0:
        plt.subplot(1, 4, i // 20 + 1)
        plt.stem(z)
```
